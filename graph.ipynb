```python
import numpy as np
import matplotlib.pyplot as plt
import argparse
import os
import csv
import pandas as pd


def F0(w):
    return -w**2

```

```python
# xiarr = np.array((1.0, 5.0, 10.0, 1000000.))
xiarr = np.array((1.0, 5.0, 1000000.))
# xiarr
figwidth = 10



plt.style.use('classic')
plt.rcParams["savefig.bbox"] = "tight"
# plt.rcParams["figure.figsize"] = (10,8)
# plt.rcParams["figure.dpi"] = 500
plt.rcParams["font.size"] = 16
# plt.rcParams["legend.frameon"] = True
# plt.rcParams["lines.linewidth"] = 5

fig, axs = plt.subplot_mosaic(

[["left column", "right top"],
["left column", "right mid"],
["left column", "right h"],
 ["left column", "right down"]], figsize=(4 * figwidth, 26)

)

colors = ['red', 'green', 'blue', 'orange']
for i in range(len(xiarr)):
    Sannikov = pd.read_csv(
        "./Figure/mercury_retirement_robust_xi={:.1f}/SannikovProblem2_xi={:.1f}_All.csv".format(xiarr[i], xiarr[i]))
    Sannikov = np.array(Sannikov)

    axs["left column"].plot(Sannikov[:, 0], Sannikov[:, 1], color=colors[i], label='$\\xi={:.0f}$'.format(xiarr[i]))
    axs["left column"].set_ylim(-1,0.15)
    axs["left column"].plot(Sannikov[:,0], F0(Sannikov[:,0]), color = 'black')
    axs["left column"].set_title("Profit $F(W)$")
    axs["left column"].grid(linestyle=':')
    axs["left column"].legend(loc='best')

    axs["right top"].plot(Sannikov[:, 0], Sannikov[:, 2],
                          color=colors[i], label='$\\xi={:.0f}$'.format(xiarr[i]))
    axs["right top"].set_ylim(0,1)
    axs["right top"].set_title("Effort $\\alpha(W)$")
    axs["right top"].grid(linestyle=':')
    axs["right top"].legend(loc='best')

    axs["right mid"].plot(Sannikov[:, 0], Sannikov[:, 3],
                          color=colors[i], label='$\\xi={:.0f}$'.format(xiarr[i]))
    axs["right mid"].set_ylim(0, 1)
    axs["right mid"].set_title("Consumption $\\pi(W)$")
    axs["right mid"].grid(linestyle=':')
    axs["right mid"].legend(loc='best')

    axs["right h"].plot(Sannikov[:, 0], Sannikov[:, 4],
                        color=colors[i], label='$\\xi={:.0f}$'.format(xiarr[i]))
    axs["right h"].set_ylim(-1,1)
    axs["right h"].set_title("Distortion $h(W)$")
    axs["right h"].grid(linestyle=':')
    axs["right h"].legend(loc='best')

    axs["right down"].plot(Sannikov[:, 0], Sannikov[:, 5],
                           color=colors[i], label='$\\xi={:.0f}$'.format(xiarr[i]))
    axs["right down"].set_ylim(-0.1, 0.1)
    axs["right down"].set_title("Distorted Drift of $W$")
    axs["right down"].grid(linestyle=':')
    axs["right down"].legend(loc='best')

# plt.savefig('CompositeSannikovRobust_All4.pdf')
plt.savefig('CompositeSannikovRobust_All3.pdf')
```
